### Using EDA(Exploratory Data Analysis)

```python
colors= sns.color_palette('PuRd_r')
labels=df2['Region'].dropna().unique()
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.title('Region_Percentage')
plt.pie(df2['Region'].value_counts(),labels=labels,colors=colors, autopct='%.2f%%')
```

#### so from above chart we can clearly see the percentage of the regions

## Region 1

```python
plt.figure(figsize=(30,8))
sns.countplot(x='Region.1',data= df2)
plt.grid(True)
plt.show
```
